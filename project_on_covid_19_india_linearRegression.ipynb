```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
import numpy as np
```

```python
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
# keepling only required columns
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
#renaming columns names
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df.head()
```

```python
df.tail()
```

```python
#current date
today=df[df.date=='2020-06-22']
```

```python
today
```

```python
max_confirmed_cases=today.sort_values(by="confirmed",ascending=False)
max_confirmed_cases
```

```python
top_state_confirmed=max_confirmed_cases[0:5]
top_state_confirmed
```

```python
sns.barplot(x='state', y='confirmed',data=top_state_confirmed,hue='state')
plt.title('Bar plot of state vs Confirmed Cases')
```

```python
max_death_cases=today.sort_values(by="deaths",ascending=False)
max_death_cases
```

```python
top_state_death=max_death_cases[0:5]
top_state_death
```

```python
sns.barplot(x='state',y='deaths',data=top_state_death,hue='state')
plt.title('Bar plot of state vs deaths')
```

```python
max_cured_cases=today.sort_values(by="cured",ascending=False)
max_cured_cases
```

```python
max_cured_state=max_cured_cases[0:5]
max_cured_state
```

```python
sns.barplot(x='state',y='cured',data=max_cured_state,hue='state')
plt.title('Bar plot of state vs cured')
```
